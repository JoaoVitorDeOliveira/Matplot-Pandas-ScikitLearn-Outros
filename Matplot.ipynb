```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot((1, 3, 5, 4), (2, 8, 7, 3))
plt.show()
```

```python
x = [1, 4, 7]
y = [2, 6, 3]
```

```python
plt.plot(x, y)
plt.xlabel('Valores da variável x')
plt.ylabel('Valores da variável y')
plt.title("Gráfico do Exercício")
plt.show()
```

```python
x2 = [1, 2, 3]
y2 = [11, 12, 15]
```

```python
plt.plot(x2, y2, label = 'Primeira Linha')
plt.legend()
plt.title('Gráfico Novo')
plt.show()
```

```python
x = [1, 4, 9, 22]
y = [3, 1, 9, 5]
```

```python
plt.bar(x, y, label = "Barras do primeiro gráfico", color = 'r')
plt.bar(x2, y2, label = "Barras do segundo gráfico", color = 'b')
plt.legend()
plt.show()
```

```python
plt.bar(x,y, label = 'Unica barra', color = 'g')
plt.legend()
plt.show()
```
